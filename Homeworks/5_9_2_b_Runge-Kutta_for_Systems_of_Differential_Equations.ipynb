# Runge-Kutta Method for Systems of Differential Equations

```python
import pandas as pd
import numpy as np
```

```python
def rk(f,a,b,h,alphas,N):
    m=2
    
    h=(b-a)/N
    t=a
    w=[[]for _x in range(m)]
    k=[[[]for _x in range(m)] for _ in range(4)]
    result=[]
    for j in range(m):
        w[j]=alphas[j]
    result.append(np.concatenate(([t],w.copy())))
    for i in range(1,N+1):
        for j in range(m):
            k[0][j]=h*f[j](t,w)
        w1=[[]for _x in range(m)]
        for _i in range(m):
            w1[_i]=w[_i]+k[0][_i]/2.0

        for j in range(m):
            k[1][j]=h*f[j](t+h/2.0,w1)
        w2=[[]for _x in range(m)]
        for _i in range(m):
            w2[_i]=w[_i]+k[1][_i]/2.0
        for j in range(m):
            k[2][j]=h*f[j](t+h/2.0,w2)
        w3=[[]for _x in range(m)]
        for _i in range(m):
            w3[_i]=w[_i]+k[2][_i]
        for j in range(m):
            k[3][j]=h*f[j](t+h/2.0,w3)
        for j in range(m):
            w[j]=w[j]+(k[0][j]+2*k[1][j]+2*k[2][j]+k[3][j])/6.0
        t=a+i*h
        result.append(np.concatenate(([t],w.copy())))
    
    return result
```

## define inputs

```python
def f1(t,w):
    u1=w[0]
    u2=w[1]
    return u1/9.0-2.0/3*u2-t**2/9.0+2.0/3
def f2(t,w):
    u1=w[0]
    u2=w[1]
    return u2+3*t-4

m=2
f=[f1,f2]
a=0
b=2
h=0.2
alphas=[-3,5]
N=10




# def f1(t,w):
#     u1=w[0]
#     u2=w[1]
#     return u2
# def f2(t,w):
#     u1=w[0]
#     u2=w[1]
#     return np.exp(2*t)*np.sin(t)-2*u1+2*u2
# a=0
# b=1
# h=0.1
# alphas=[-0.4,-0.6]
```

## Actual Results

```python
u1=lambda t:-3*np.exp(t)+t**2
u2=lambda t:4*np.exp(t)-3*t+1
h=(b-a)/N
result_actual=[]
for i in np.arange(a,b+h,h):
    result_actual.append([u1(i),u2(i)])
df_actual=pd.DataFrame(result_actual,columns=['u0_actual','u1_actual'])
```
